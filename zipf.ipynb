```python
import nltk
nltk.download('punkt')

import matplotlib.pyplot as plt
import math
```

```python
f = open('../kantuoft_critiqueofjudgem.txt')
text = f.read()
f.close()
```

```python
words = nltk.word_tokenize(text)

word_freq = nltk.FreqDist(words).most_common()

indexes = list(range(1, len(word_freq) + 1))
frequences = list(map(lambda pair: pair[1], word_freq))
```

```python
plt.plot(indexes, frequences, '.')
plt.suptitle('Линейный график')
plt.show()
```

```python
plt.plot(list(map(math.log, indexes)), list(map(math.log, frequences)), '.')
plt.suptitle('Логарифмический график')
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(np.log(np.array(indexes).reshape(-1,1)), np.log(np.array(frequences)))

print(f'k={reg.coef_} b={reg.intercept_}')


def line(x):
    return reg.coef_[0] * x + reg.intercept_


plt.plot(list(map(math.log, indexes)), list(map(math.log, frequences)), '.',
    [0, 10],
    [line(0), line(10)],
    '--'
)
plt.suptitle('Логарифмический график')
plt.show()
```
